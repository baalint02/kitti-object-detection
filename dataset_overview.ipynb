```python
from kitti_detection import config
from kitti_detection.dataset import KittiDetectionDataset, load_train_val_test_dataset
from kitti_detection.utils import display_samples_v

from torch.utils.data import DataLoader
```

```python
dataset = KittiDetectionDataset(image_dir_path=config.DATA_IMAGE_DIR_PATH, label_dir_path=config.DATA_LABEL_DIR_PATH)
```

```python
dataset[0]
```

```python
data_loader = DataLoader(dataset, batch_size=4, shuffle=True, collate_fn=lambda batch: tuple(batch))
```

```python
display_samples_v(next(iter(data_loader)))
```
